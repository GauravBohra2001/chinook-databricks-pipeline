```python
%sql
USE CATALOG chinook_catalog;
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS chinook_catalog.pipeline_metadata;

CREATE TABLE IF NOT EXISTS chinook_catalog.pipeline_metadata.parent_metadata (
    table_name    STRING,
    file_name     STRING,
    active_flag   STRING,
    created_date  DATE,
    modified_date DATE
);
```

```python
%sql
CREATE TABLE IF NOT EXISTS chinook_catalog.pipeline_metadata.child_metrics_metadata (
    table_name        STRING,
    execution_time    TIMESTAMP,
    status            STRING,
    source_row_count  LONG,
    target_row_count  LONG,
    file_location     STRING,
    created_date      DATE
);
```
